**Q.1**

Given:
$$\vec{B} = B_o \hat{y}$$
$$\vec{g} = -g \hat{z}$$

We can define the velocity of the object along the three components of motion as well:
$$\vec{v} = \left<v_x, v_y, v_z\right>$$

We also know that the object is placed at rest at the origin, at $t = 0$. 

Using the formula for the magnetic force:

$$\vec{F_B} = q \left(\vec{v}\: \times\: \vec{B}\right) = \begin{vmatrix} \hat{x} & \hat{y} & \hat{z} \\ v_x & v_y & v_z \\ 0 & B_o & 0 \end{vmatrix}$$

$$\vec{F_B} = q \left[-v_z B_o \hat{x} + v_x B_o \hat{z}\right]$$

Creating our Equations of Motion with the given information:

$m\dot{v_y} = 0$

$m\dot{v_x} = - q v_z B_o$

$m\dot{v_z} = q v_x B_o - m g$

Simplifying, we get:

$$\dot{v_x} = - \frac{q B_o}{m} v_z$$

$$\dot{v_z} =  \frac{q B_o}{m} v_x - g$$

Let us define a constant $k = -\frac{q B_o}{m}$ to get $\dot{v_x} = k v_z$ and $\dot{v_z} = -k \left(-v_x + \frac{g}{k}\right)$. 

We also define a complex number $\eta (t) = v_z + i (-v_x + g/k)$. 

Since $g$ and $k$ are constants, when we differentiate $\eta(t)$ with respect to $t$:

$$\dot{\eta}(t) = \dot{v_z} + i \left(-\dot{v_x} + \dot{(g/k)}\right) = \dot{v_z} - i \dot{v_x}$$
$$= -k \left(-v_x + g/k\right) - i k v_z$$
$$= -k \left[\left(-v_x + g/k\right) - i v_z\right]$$

Using the complex property that $a + ib = - (b + ai)$, we can simplify the equation above to:

$$\dot{\eta}(t) = -i k \eta(t)$$

Using the method of Separation of Variables:

$$\int\frac{1}{\eta} d\eta = \int -ik\: dt$$

$$\ln{\eta} - \ln{C} = -ikt$$

Where $C$ is a constant of integration we decided to take on the left hand side. This gives us an equation for $\eta(t)$:

$$\eta(t) = C e^{-ikt}$$

To solve for this constant of integration, we need to first identify the equation we have, and then make an educated guess. Let us guess that $C = ce^{i\delta}$. We then get:

$$\eta(t) = c e^{-ikt + i\delta} = c e^{-i\left(kt - \delta \right)}$$

Using Euler's Expansion, we get:

$$\eta(t) = c \cos{\left(kt - \delta\right)} - i\: c \sin{\left(kt - \delta\right)}$$

We now have a real and an imaginary component that we can compare our initial values to. 

$$v_z = c \cos{\left(kt - \delta\right)}$$

$$\left(-v_x + g/k\right) = - c \sin{\left(kt - \delta\right)}$$

We integrate both equations with repect to $t$ to find $x(t)$ and $z(t)$:

$$x(t) = \int c \sin{\left(kt - \delta\right)} dt + \int g/k\: dt = \frac{-c}{k} \cos{\left(kt - \delta\right)} + \frac{gt}{k} + c_x$$

$$z(t) = \int c \cos{\left(kt - \delta\right)} dt = \frac{c}{k} \sin{\left(kt - \delta\right)} + c_z$$

We are given initial conditions that the object is at rest at the origin at $t = 0$. This tells us that at $t = 0$, $v_y = 0, c_x = c_y = c_z = 0$. 

Summarizing the data we have:

$$\boxed{x(t) = \frac{-c}{k} \cos{\left(kt - \delta\right)} + \frac{gt}{k}}$$

$$\boxed{y(t) = 0}$$

$$\boxed{z(t) = \frac{c}{k} \sin{\left(kt - \delta\right)}}$$

Where $k = \frac{q B_o}{m}$ is the angular frequency of a "cycloid".

**Q.2**

We will be using cylindrical coordinates to set this problem up. We also must define some variables that we will be using in this setup.
$r = l \sin{\theta}$, where $r$ is the radius of the circle traced by the tetherball when in uniform circular motion. $l$ is the length of the string of the tetherball, and $\theta$ is the angle subtended by the tetherball with respect to the vertical. 

The ball has two forces acting on it - the gravitational acceleration downward, and the tension of the string toward the point of tethering. In coordintes, the force equation can be described as:

$$F_{ball} = m\vec{a} = -T \sin{\theta}\: \hat{r} + T \cos{\theta}\: \hat{z} - mg \hat{z}$$

The acceleration for an object in cylindrical coordinates is given by:

$$\vec{a} = \left(\ddot{r} - r\dot{\phi}^2\right) \hat{r} + \left(r \ddot{\phi} + 2\dot{r}\dot{\phi}\right) \hat{\phi} + \left(\ddot{z}\right) \hat{z}$$

Comparing the two equations:

$m \left(\ddot{r} - r\dot{\phi}^2\right) = - T \sin{\theta}$

$m \left(r \ddot{\phi} + 2\dot{r}\dot{\phi}\right) = 0$

$m \ddot{z} = T \cos{\theta} - mg$

Since $\theta$ is the vertical angle, and it must remain constant, $\ddot{r} = \dot{r} = 0$. Since $z = l \cos{\theta}$ is a constant, that means that $\ddot{z} = \dot{z} = 0$. This simplifies our equations to be:

$T \cos{\theta} = mg \rightarrow T = \frac{mg}{\cos{\theta}}$

$m r \dot{\phi}^2 = T \sin{\theta}$

$m r \ddot{\phi} = 0$

Combining the first and second equations:

$$m r \dot{\phi}^2 = \frac{mg}{\cos{\theta}} \sin{\theta} = m \left(l \sin{\theta}\right) \dot{\phi}^2 \rightarrow \dot{\phi} = \sqrt{\frac{g}{l \cos{\theta}}}$$

**2.(i)**

$$\theta = \cos^{-1}{\left(\frac{mg}{T}\right)} = \cos^{-1}{\left(\frac{g}{l \dot{\phi}^2}\right)} = \sin^{-1}{\left(\frac{T}{l}\right)}$$ 

**2.(ii)**

$$\dot{\phi} = \sqrt{\frac{g}{l \cos{\theta}}}$$

As $\dot{\phi} \rightarrow 0, \theta \rightarrow 0$ such that $\dot{\phi} = \sqrt{\frac{g}{l \cos{\theta}}} \approx \sqrt{\frac{g}{l}}$. 

The frequency is similar to that of a pendulum, however with different constants.

**Q.3**

At the moment the ball leaves the sphere, we define the following variables:
$h$ is the height from the center of the sphere. $R - h$ is the distance from the top of the sphere to the point of leaving. 
$\theta$ is the angle subtended from the vertical axis. $v_o$ is the initial horizontal velocity. $N$ is the normal force on the object.

Since the gravitational force only acts along the z-axis, the component of the gravitational force along the ball in the $\hat{r}$ direction is $mg \cos{\theta}$.

So the force along the $\hat{r}$ direction is:

$$mg \cos{\theta} - N = \frac{m v^2}{R} \tag{1}$$

Where $v \neq v_o$. 

Using the Conservation of Energy: $1/2 m v^2 - mgh = 0$, we get an equation in terms of $R$:

$$\frac{1}{2}m v^2 - m g \left(R - R \cos{\theta}\right) = 0$$

$$g R \left(1 - \cos{\theta}\right) = \frac{1}{2}\left(v^2 - {v_o}^2\right) \tag{2}$$

Putting eqs (1) and (2) together, and the fact that $N = 0$ at the point the ball leaves to solve for $h = R - R \cos{\theta}$:

$$2 g \left(R - R\cos{\theta}\right) = g R \cos{\theta} - {v_o}^2$$

$$\cos{\theta} = \frac{{v_o}^2}{gR} + 2 - 2 \cos{\theta}$$

$$3 \cos{\theta} = 2 + \frac{{v_o}^2}{gR}$$

$$\cos{\theta} = \frac{2}{3} + \frac{{v_o}^2}{3gR}$$

$$h = R - R \cos{\theta} = R \left[1 - \frac{2}{3} - \frac{{v_o}^2}{3gR}\right]$$

$$\boxed{h = \frac{R}{3}\left[1 - \frac{{v_o}^2}{gR}\right]}$$

The angle at which the object loses contact:

$$\boxed{\theta = \cos^{-1}{\left(\frac{2}{3} + \frac{{v_o}^2}{3gR}\right)}}$$

This angle depends on the value of $v_o$. If $v_o$ is infinitesimal, then $\theta$ can be approximated to about $48.19^\circ$ from the vertical axis.
