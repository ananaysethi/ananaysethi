$$\vec{F} = F_o\: e^{(-t/t_c)}\: \hat{x}$$

We are given, that at $t = 0$, $x = 0,\: \dot{x} = v_0$

Since we are working only in one dimension, we can remove the vectors for calculation purposes. This gives us our new equation:

$$m\ddot{x} = F_o\: e^{(-t/t_c)} \rightarrow \ddot{x} = \frac{F_o}{m}\: e^{(-t/t_c)}$$

Integrating both sides with respect to $t$:

$$\int_{v_0}^{v} \frac{dv'}{dt} dt = \frac{F_o}{m} \int_{0}^{t}e^{-t/t_c} dt$$

(Note that $v'$ here has only been used as a placeholder term)

$$v - v_0 = \frac{F_o\: t_c}{m} \left[1 - e^{-t/t_c}\right] \rightarrow v(t) = \frac{F_o\: t_c}{m} \left[1 - e^{-t/t_c}\right] + v_0$$

Now we integrate the velocity with respect to $t$ once more to get the displacement:

$$\int_{0}^{x} \frac{dx'}{dt} dt = \frac{F_o\: t_c}{m} \int_{0}^{t}1 - e^{-t/t_c}\: dt + \int_{0}^{t}v_0\: dt$$

This gives us:

$$x(t) = \frac{F_o\: t_c}{m} \left[t + t_c\:e^{-t/t_c} - t_c\right] + v_0 t$$

In simpler terms:
$$\boxed{x(t) = \frac{F_o\: {t_c}^2}{m} \left[e^{-t/t_c} - 1\right] + \frac{F_o\: t_c\: t}{m} + v_0 t}$$

$$\vec{F} = - F_o e^{-x / x_c}\: \hat{x}$$

Here, the force is a function of position, and we are given the following initial conditions:

At $t = 0$, $x = 0$, and $v = v_0$.

$$\ddot{x} = \frac{d}{dt}(\dot{x}) = \frac{dx}{dt} \frac{d\dot{x}}{dx} = \dot{x}\frac{d\dot{x}}{dx} = v\frac{dv}{dx}$$

This gives us an equation for the force:

$$mv\frac{dv}{dx} = - F_o e^{-x/x_c}$$

$$\int{v} dv = -\frac{F_o}{m} \int{e^-x/x_c} dx$$

$$\frac{v^2}{2} = -\frac{F_o}{m} (- x_c e^{-x/x_c} + c_o)$$

Where $c_o$ is a constant of integration. This can be further simplified to:

$$\frac{v^2}{2} = \frac{F_o}{m} (x_c e^{-x/x_c}) - \frac{F_o c_o}{m}$$

Using the initial conditions, where $x = 0$ and $v = v_o$ at $t = 0$:

$$\frac{{v_o}^2}{2} = \frac{F_o}{m}x_c - \frac{F_o}{m}c_o$$

$$c_o = x_c - \frac{m {v_o}^2}{2 F_o}$$

We can rewrite the equation for the velocity of this force:

$$\dot{x} = v = \sqrt{\frac{2 F_o}{m} (x_c e^{-x/x_c} - x_c + \frac{m {v_o}^2}{2 F_o})}$$

Simplifying for the $v_o$ value:

$$v(x) = \sqrt{\frac{- 2 F_o}{m}x_c (1 - e^{-x/x_c}) + {v_o}^2}$$

We know that $v = \dot{x} = \frac{dx}{dt}$

Using the Method of Separation of Variables, we get:

$$dt = \frac{dx}{\sqrt{\frac{- 2 F_o}{m}x_c (1 - e^{-x/x_c}) + {v_o}^2}}$$

Integrating both sides, we get a function for $t(x)$:

$$\int{dt} = \int{\frac{dx}{\sqrt{\frac{- 2 F_o}{m}x_c (1 - e^{-x/x_c}) + {v_o}^2}}}$$

$$\boxed{t(x) = \int_{0}^{x}{\frac{dx}{\sqrt{{v_o}^2 - \frac{2 F_o}{m}x_c (1 - e^{-x/x_c})}}}}$$


There is no analytical solution to this integral, since the solution depends on the value of ${v_o}^2$, specifically $v_o$. If ${v_o}^2 > \frac{2 F_o}{m}x_c (1 - e^{-x/x_c})$, then we have an analytical solution for the integral that is real. For ${v_o}^2 \leq \frac{2 F_o}{m}x_c (1 - e^{-x/x_c})$, we will not be able to calculate a real solution.

Below is an attempt at an analytical solution using sympy in Python. It returns the same integral back, signifying that there exists no unconditional analytical solution.

```python
import sympy as sy

F_o, x_c, x, m, v_o = sy.symbols('F_o, x_c, x, m, v_o')

sy.integrate(1/sy.sqrt((2*F_o*x_c/m) * (sy.exp(-x/x_c) - 1) - v_o**2), x)
```

Plotting Taylor Series Expansions upto the third order for the following functions:

**3.(a) - $sin(x)$ centered at $(x = 0.25)$**

$$sin(x) \approx sin(0.25) + \frac{1}{1!}\frac{d}{dx}sin(x)|_{sin(x) = sin(0.25)}(x - 0.25)^1 + \frac{1}{2!}\frac{d^2}{dx^2}sin(x)|_{sin(x) = sin(0.25)}(x - 0.25)^2 + \frac{1}{3!}\frac{d^3}{dx^3}sin(x)|_{sin(x) = sin(0.25)}(x - 0.25)^3 ...$$

Simplifying, we get:

$$sin(x) \approx 0.198669 + 0.980067 (x - 0.25) - 0.09933 (x - 0.25)^2 - 0.16334 (x - 0.25)^3$$

We shall now plot the original function, its first, second and third order approximations respectively. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: np.sin(x)
f1 = lambda x: np.sin(0.25) + np.cos(0.25)*(x - 0.25)
f2 = lambda x: np.sin(0.25) + np.cos(0.25)*(x - 0.25) - 0.5*np.sin(0.25)*(x - 0.25)**2
f3 = lambda x: np.sin(0.25) + np.cos(0.25)*(x - 0.25) - 0.5*np.sin(0.25)*(x - 0.25)**2 - (1/6)*np.cos(0.25)*(x - 0.25)**3

x = np.linspace(0.01, 2)
plt.plot(x, f(x), label = 'Original Function')
plt.plot(x, f1(x), label = 'First Approximation')
plt.plot(x, f2(x), label = 'Second Approximation')
plt.plot(x, f3(x), label = 'Third Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximations for sin(x), center = 0.25')
plt.grid()
plt.legend()
plt.show()
```

**4.(a)**

$$U(x) = x^{-6}\: e^{x^2}$$

To find the postion where the potential energy is a minimum, we must employ the Second Derivative Test. 

Finding the first derivative:

$$\frac{dU}{dx} = -6x^{-7}e^{x^2} + 2x^{-5}e^{x^2} = x^{-7}\cdot e^{x^2}\left[-6 + 2x^2\right]$$

At the minima, the first derivative is equal to zero:

$$x^{-7}\cdot e^{x^2}\left[-6 + 2x^2\right] = 0$$

which gives us $x = 0$, or $x = \pm \sqrt{3}$ as the critical points.

Now, finding the second derivative:

$$\frac{d^2 U}{dx^2} = x^{-8} e^{x^2} \left[4x^4 - 22x^2 + 42\right]$$

We have been given the condition that we are looking at the $x > 0$ realm, so we plug in $x = \sqrt{3}$ into the second derivative:

$$\frac{d^2 U}{dx^2}|_{x = \sqrt{3}} = \frac{4e^3}{27}$$

Since $\frac{4e^3}{27} > 0$, we can definitively say that the minimum potential energy exists at $x = \sqrt{3}$.

**4.(b)**

To look at small oscillations around the minimum, we need to look at the Taylor Series of the Potential Energy Function:

$$U(x) \approx U(x_o) + \frac{dU}{dx}|_{x = x_o} (x - x_o) + \frac{1}{2!}\frac{d^2 U}{dx^2}|_{x = x_o} (x - x_o)^2 ...$$

where $x_o$ is the value we calculated in the previous part of the question, i.e. $x_o = \sqrt{3}$. 

We see that the first term of the approximation is a constant term, and the first derivative is zero (since we are analysing the function at the minimum), and most terms after the second order approximation are assumed to be insignificant for this calculation. Therefore, after simplifying, we get:

$$U(x) \approx \frac{1}{2}\frac{d^2 U}{dx^2}|_{x = x_o} (x - x_o)^2$$

For a spring in harmonic oscillations, the potential energy function could be described as 

$$U(x) = \frac{1}{2} k \left(x - x_o\right)^2$$

where $k = m\:\omega^2$ is the spring constant. 

Equating the two expressions, we get:

$$m\:\omega^2 = \frac{d^2 U}{dx^2}|_{x = x_o}$$

which gives us a value for the angular frequency $\omega$:

$$\omega = \sqrt{\frac{1}{m} \frac{d^2 U}{dx^2}|_{x = x_o}} = \sqrt{\frac{4e^3}{27m}}$$

$$\boxed{\omega = \sqrt{\frac{4e^3}{27m}}}$$
