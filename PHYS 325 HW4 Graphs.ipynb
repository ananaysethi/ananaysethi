```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.constants as const
import astropy.units as u
```

```python
m1 = 1 * const.M_earth
m2 = 1000 * u.kg

R = 2 * const.R_earth
r = np.linspace(R, 2*R, 100000)

v_circ = np.sqrt(const.G * m1 / R)
l = m2 * v_circ * r
l_new = 0.7 * l

U_eff = (l**2 / (2 * m2 * r**2)) - (const.G * m1 * m2 / r)
U_eff_new = (l_new**2 / (2 * m2 * r**2)) - (const.G * m1 * m2 / r)

fig = plt.figure(figsize=(10,6))

plt.plot(r, U_eff, label='Original Effective Potential')
plt.plot(r, U_eff_new, label='New Effective Potential')
plt.ylabel(r'Effective Potential $[\times 10^{10} J]$')
plt.xlabel(r'The Distance from the Center of the Earth to Satellite$[\times 10^{7} m]$')
plt.title('Effective potential (starting from the surface of the Earth since potential inside is different)')
plt.grid()
plt.legend()
plt.show()
```
