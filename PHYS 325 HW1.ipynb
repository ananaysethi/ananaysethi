```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.sin(x)
f1 = lambda x: np.sin(0.2) + np.cos(0.2)*(x - 0.2)
f2 = lambda x: np.sin(0.2) + np.cos(0.2)*(x - 0.2) - 0.5*np.sin(0.2)*(x - 0.2)**2
f3 = lambda x: np.sin(0.2) + np.cos(0.2)*(x - 0.2) - 0.5*np.sin(0.2)*(x - 0.2)**2 - (1/6)*np.cos(0.2)*(x - 0.2)**3
```

```python
x = np.linspace(0.01, 2)
plt.plot(x, f(x), label = 'Original Function')
plt.plot(x, f1(x), label = 'First Approximation')
plt.plot(x, f2(x), label = 'Second Approximation')
plt.plot(x, f3(x), label = 'Third Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximations for sin(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
g = lambda x: np.log(x)
g1 = lambda x: np.log(0.2) + (1/0.2)*(x - 0.2)
g2 = lambda x: np.log(0.2) + (1/0.2)*(x - 0.2) - (1/0.08)*(x - 0.2)**2
g3 = lambda x: np.log(0.2) + (1/0.2)*(x - 0.2) - (1/0.08)*(x - 0.2)**2 + (2/0.048)*(x - 0.2)**3
```

```python
x = np.linspace(0.01, 2)
plt.plot(x, g(x), label = 'Original Function')
plt.plot(x, g1(x), label = 'First Approximation')
plt.plot(x, g2(x), label = 'Second Approximation')
plt.plot(x, g3(x), label = 'Third Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximations for ln(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
h = lambda x: np.exp(x)
h1 = lambda x: np.exp(0.2) + np.exp(x)*(x - 0.2)
h2 = lambda x: np.exp(0.2) + np.exp(x)*(x - 0.2) + (1/2)*np.exp(x)*(x - 0.2)**2
h3 = lambda x: np.exp(0.2) + np.exp(x)*(x - 0.2) + (1/2)*np.exp(x)*(x - 0.2)**2 + (1/6)*np.exp(x)*(x - 0.2)**3
```

```python
x = np.linspace(0.01, 2)
plt.plot(x, h(x), label = 'Original Function')
plt.plot(x, h1(x), label = 'First Approximation')
plt.plot(x, h2(x), label = 'Second Approximation')
plt.plot(x, h3(x), label = 'Third Approximation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Series Approximations for e^x')
plt.grid()
plt.legend()
plt.show()
```
