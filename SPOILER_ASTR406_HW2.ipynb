```python
# B - V
B_V_MS = ([
    -0.32, -0.32, -0.32, -0.30, -0.18, -0.14, -0.11, 0.0, 0.19, 0.32, 0.41,
    0.59, 0.65, 0.69, 0.84, 1.08, 1.32, 1.41, 1.5, 1.6
])
# absolute magnitude
M_V_MS = ([
    -5.8, -5.2, -4.3, -3.7, -1.4, - 1.0, -0.25, 0.8, 1.8, 2.4, 3.3, 4.2, 4.83,
    4.93, 5.9, 7.5, 8.3, 8.9, 11.2, 12.7
])

plt.plot(B_V_MS, M_V_MS)

plt.plot(0.65, 4.83, 'ro', label='Sun')

plt.gca().invert_yaxis()

plt.xlabel('B - V color')
plt.ylabel('Absolute Magnitude')
plt.title('Absolute Magnitude vs. B - V')
plt.legend()
```

## b)

$$
B-V = 0.5, \space m_V = 8
$$

```python
m_V_star = 8
B_V_star = 0.5

M_V_star = np.interp(B_V_star, B_V_MS, M_V_MS)
M_V_star
```

```python
print("The absolute magnitude of the star is:", "{:.2e}".format(M_V_star))
```

$$ 
\text{distance modulus:} \space m_V - M_V = 5 \log d - 5 \\
d=10^{\frac{m_V-M_V+5}{5}}
$$

```python
distance_star = 10**((m_V_star - M_V_star + 5) / 5)
distance_star # [pc]
```

```python
print("The distance of the star is:", "{:.2e}".format(distance_star), "pc")
```

## c)

$$ 
d=10^{\frac{m_V-M_V+5-A_V}{5}}
$$

We can expect the true distance to be smaller than what we got.

```python
A_V = 0.4

distance_star_actual = 10**((m_V_star - M_V_star + 5 - A_V) / 5)
distance_star_actual # [pc]
```

```python
print("The actual distance of the star is:", "{:.2e}".format(distance_star_actual), "pc")
```
